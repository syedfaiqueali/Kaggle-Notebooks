# Imports

```python
import os
import warnings
import pandas as pd
import numpy as np

from langchain.agents.agent_types import AgentType
from langchain_experimental.agents.agent_toolkits import create_csv_agent
from langchain_openai import ChatOpenAI, OpenAI

warnings.simplefilter('ignore')
```

# Constant

```python
DATASET_PATH = '/kaggle/input/stock-exchange-data/indexProcessed.csv'
os.environ["OPENAI_API_KEY"] = '<KEY WILL GO HERE>' # Remove it later
```

# Data Exploration Using AI CSV Agent

```python
agent = create_csv_agent(
    ChatOpenAI(temperature=0, model="gpt-3.5-turbo-0613"),
    DATASET_PATH,
    verbose=True,
    agent_type=AgentType.OPENAI_FUNCTIONS,
)
```

```python
agent.run("how many rows are there?")
```

```python
agent.run("how many column are there")
```

```python
agent.run("What are the columns types?")
```

```python
agent.run("What is the 'High' mean")
```

```python
agent.run("What is the 'Close' mean")
```

```python
agent.run("What is the max closing value?")
```

```python
agent.run("What is the max closing USD value?")
```

```python
agent.run("Which particular record hai highest USD value?")
```

```python
agent.run("Does the data follows normal distribution?")
```
