```python
import numpy as np
import matplotlib.pyplot as plt
```

# Utils

```python
def draw_concentric_circles(inner_radius, coping_width=4):
    # Define radii of the circles
    outer_radius = inner_radius + 12
    
    # Create a figure and axis to plot
    fig, ax = plt.subplots()
    ax.set_aspect('equal')  # Keep the aspect ratio of the plot as 1:1

    # Plot the inner circle
    inner_circle = plt.Circle((0, 0), inner_radius, color='blue', fill=False, linewidth=2)
    ax.add_artist(inner_circle)

    # Plot the outer circle
    outer_circle = plt.Circle((0, 0), outer_radius, color='green', fill=False, linewidth=2)
    ax.add_artist(outer_circle)

    # Draw lines from outer circle to center
    circumference = 2 * np.pi * outer_radius  # Calculate circumference of outer circle
    num_lines = int(circumference // coping_width)  # Number of lines with coping width spacing along circumference

    # Define angles where the lines will be drawn (spaced equally)
    angles = np.linspace(0, 2 * np.pi, num_lines, endpoint=False)

    # Draw each line
    for angle in angles:
        # Calculate the end point of the line on the outer circle
        x_outer = outer_radius * np.cos(angle)
        y_outer = outer_radius * np.sin(angle)
        
        # Draw a line from center (0,0) to the outer point (x_outer, y_outer)
        ax.plot([0, x_outer], [0, y_outer], color='red', linewidth=1)

    # Set limits for the plot
    ax.set_xlim([-outer_radius - 2, outer_radius + 2])
    ax.set_ylim([-outer_radius - 2, outer_radius + 2])
    
    # Show grid for better visualization
    ax.grid(True)
    
    # Show plot
    plt.title('Concentric Circles with Lines')
    plt.show()
```

# Code

```python
# Example usage
inner_radius = 0
coping_width = 4
draw_concentric_circles(inner_radius, coping_width)
```

```python
# Example usage
inner_radius = 12
coping_width = 4
draw_concentric_circles(inner_radius, coping_width)
```
